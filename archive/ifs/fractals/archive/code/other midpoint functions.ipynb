```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline

def deg2rad(deg):
    return deg *np.pi/180
```

```python
initial_points = [[0,1],
                  [np.cos(deg2rad(30)),-np.sin(deg2rad(30))],
                  [-np.cos(deg2rad(30)),-np.sin(deg2rad(30))]]
```

```python
def get_midpoint(p1,p2,jump):
    x1 = (((p2[0]-p1[0])*jump)+p1[0])
    y1 = (((p2[1]-p1[1])*jump)+p1[1])
    return [x1,y1]
```

```python
def get_midpoint(p1,p2,jump):
    x1 = (((p2[0]-p1[0])*jump)+p1[0])
    y1 = (((p2[1]-p1[1])*jump)+p1[1])
    
    return y1, x1
```

```python
def plot_game(initial_points,num_points,jump,alpha,s):
    current_point = [0,0]

    plt.figure(figsize=(8,7))
    for point in initial_points:
        plt.scatter(point[0],point[1],c='red')
    plt.scatter(current_point[0],current_point[1],c='blue')

    x_points = []
    y_points = []
    for i in range(num_points):
        current_point = get_midpoint(current_point,random.choice(initial_points),jump)
        x_points.append(current_point[0])
        y_points.append(current_point[1])

    plt.scatter(x_points,y_points,c='blue',alpha=alpha,s=s)
```

```python
def get_midpoint(p1,p2,jump):
    x1 = (np.sin(p1[0]) + np.sin(p2[0]))/(1/jump)
    y1 = (np.sin(p1[1]) + np.sin(p2[1]))/(1/jump)
    
    return y1,x1
```

```python
plot_game(initial_points,10000,0.6,0.1,0.5)
```
